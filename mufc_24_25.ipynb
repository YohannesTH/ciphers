```python
import matplotlib.pyplot as plt

def draw_pitch():
    """Draws a vertical football pitch using matplotlib"""
    fig, ax = plt.subplots(figsize=(7, 10))
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.axis('off')

    # Outer boundary
    plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], color='black')

    # Center circle
    center_circle = plt.Circle((50, 50), 10, color='black', fill=False)
    ax.add_patch(center_circle)

    # Halfway line
    plt.plot([0, 100], [50, 50], color='black')

    # Penalty areas
    plt.plot([0, 18, 18, 0], [13, 13, 87, 87], color='black')
    plt.plot([100, 82, 82, 100], [13, 13, 87, 87], color='black')

    return fig, ax

def plot_players(ax, players):
    """Plots the players on the pitch"""
    for pos, data in players.items():
        x, y = data['coords']
        ax.text(x, y, pos, fontsize=10, ha='center', color='red', fontweight='bold')
        for i, player in enumerate(data['players']):
            ax.text(x, y - (i+1)*3, player, fontsize=8, ha='center', color='black')

def main():
    fig, ax = draw_pitch()

    players = {
        'CF': {'coords': (50, 90), 'players': ['R. Hojlund +3*', 'J. Zirkzee +4*']},
        'LW': {'coords': (20, 80), 'players': ['A. Garnacho +4', 'LW +4' ]},
        'RW': {'coords': (80, 80), 'players': ['A. Diallo +5', 'RW +4' ]},
        'AM': {'coords': (50, 80), 'players': ['B. Fernandes +2*', 'M. Mount +3*' ]},
        'CM': {'coords': (50, 60), 'players': ['K. Mainoo +2*', 'CM +4']},
        'DM': {'coords': (50, 40), 'players': ['M. Ugarte +4*', 'T. Collyer +2']},
        'LB': {'coords': (10, 40), 'players': ['P. Dorgu +5*', 'Shaw +2', 'H. Amass +2']},
        'RB': {'coords': (90, 40), 'players': ['D. Dalot +3*', 'N. Mazraoui +3*']},
        'LCB': {'coords': (30, 30), 'players': ['L. Martinez +2*', 'H. Maguire +1', 'Heaven +4*']},
        'RCB': {'coords': (70, 30), 'players': ['M. de Ligt +4*', 'L. Yoro +4*']},
        'GK': {'coords': (50, 10), 'players': ['A. Onana +3*', 'A. Bayindir +2*']},
    }

    plot_players(ax, players)
    plt.title("Manchester United's Squad Depth 2024-25", fontsize=14, fontweight='bold')
    plt.show()

if __name__ == "__main__":
    main()
```
